# Cleaning countries'names 

```python
def f_country() -> list:
    """Doccumentation:
    This function returns the list of countries in the world, 
    input : no value
    return : a list of contries got by using the library pycountry  
    """
    list_pays = []
    for pays in list(py.countries):
        a = re.search("name='"+'.+'+", numeric=", str(pays))
        if a:
            a = a.group(0)[6:-11]
            list_pays.append(a.lower())
    return list_pays
```

```python
def replace_elt(elt1, elt2, my_liste):
    """Doccumentation :

    This function replace elt1 by elt2 in a string ( sentence )
    Input : elt1, elt2, my_liste ( is list of sentence)
    Output : Doesn't have a return value 

    """
    for i in range(len(my_liste)):
        w = str(my_liste[i])
        my_liste[i] = re.sub(elt1, elt2, w)
```

# Cleaning keywords

```python
# Loading data which contains keywords
df_contain = pd.read_csv('tableContenir.csv')[['pii', 'Mots']]
df_contain
```

```python
key_words = df_contain['Mots'].values
key_words
```

```python
!pip install unidecode
import unidecode
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
tokenizer = nltk.RegexpTokenizer(r'\w+')
nltk.download('stopwords')
```

```python
def cleaning(listeSentences, langue='french'):
    """Doccumentation:
    This function cleans a list of sentences,for each sentence it removes stopword (like, of, that...), removes 
    words that length is less than 2, and convert to lowercase
    Input : 
          listeSentences : a list of sentence (string)
          langue : To choose the langue, the deault value is french
    Output: a list which contains the cleaned sentences 
    """

    if langue == '':
        return [' '.join([unidecode.unidecode(w) for w in tokenizer.tokenize(phrase)]).lower() for phrase in listeSentences]
    else:
        stopWords = set(stopwords.words(langue))
        sent = [' '.join([unidecode.unidecode(w) for w in (tokenizer.tokenize(str(
            phrase))) if w not in stopWords and len(w) > 2]).lower() for phrase in listeSentences]
        return [w.split() for w in sent]
```

```python
df_contain['Mots'] = cleaning(key_words, langue='english')
```

```python
df_contain
```

```python
"""Some word are compound, that means they can be formed by two subword or more. To handle them, 
if the word is formed by more than two subword we transforme it by considering each subword as a single word that
which will be assigned of course to the article of the word from which it comes

"""

n = len(df_contain)
print(n)
dico = {'pii': [], 'Mots': []}
for i in range(n):
    len_word = len(df_contain['Mots'].iloc[i])
    if len_word > 2:
        for j in range(len_word):
            dico['pii'].append(df_contain['pii'].iloc[i])
            dico['Mots'].append([df_contain['Mots'].loc[i][j]])
    else:
        dico['pii'].append(df_contain['pii'].iloc[i])
        dico['Mots'].append(df_contain['Mots'].loc[i])
```

```python
df_contain_words = pd.DataFrame.from_dict(dico)
```

```python
df_contain_words
```

```python
def joint_word(sent)-> str:
    """Doccumentation:
        This function transforms a list of string to a single string 
    
    """
    return ' '.join([w for w in sent])
```

```python
df_contain_words['Mots'] = df_contain_words['Mots'].apply(joint_word)
```

## Cleaning the title of each article

```python
# A version of the function cleaning which return a str, not a pyhton list


def cleaning_str(listeSentences, langue='french'):
    """Doccumentation:
        This function cleans a list of sentences,for each sentence it removes stopword (like, of, that...), removes 
        words that length is less than 2, and convert to lowercase
    Input : 
          listeSentences : a list of sentence (string)
          langue : To choose the langue, the deault value is french
    Output: a string  
    """

    if langue == '':
        return [' '.join([unidecode.unidecode(w) for w in tokenizer.tokenize(phrase)]).lower() for phrase in listeSentences]
    else:
        stopWords = set(stopwords.words(langue))
        return [' '.join([unidecode.unidecode(w) for w in (tokenizer.tokenize(str(phrase))) if w not in stopWords and len(w) > 2]).lower() for phrase in listeSentences]
```

```python
# geting titles in a list to aplly the function cleaning_str
art_title = df_article['titre'].values 
```

```python
df_article['titre'] = cleaning_str(art_title, langue='english')
```
